```python
import numpy as np
import matplotlib.pyplot as plt

# Define constants
V_DD = 1.5  # Supply voltage
V_M = 0.75  # Midpoint voltage
V_OH = 1.1  # Output High voltage
V_OL = 0.4  # Output Low voltage
V_IL = 0.77  # Input Low voltage
V_IH = 0.73  # Input High voltage
g = -38.09  # Slope of VTC at V_M (negative)

# Generate input voltage range around the transition region
V_in = np.linspace(0.6, 0.9, 500)  # Adjusted range to zoom in around V_M

# Define the VTC using a tanh function to simulate steep transitions
# Adjust the steepness factor (alpha) based on g to simulate the slope at V_M
alpha = 50  # This value affects the steepness; higher makes a steeper transition
V_out = V_OH + (V_OL - V_OH) / 2 * (1 + np.tanh(alpha * (V_in - V_M)))

# Calculate tangent line at V_M
V_in_tangent = np.linspace(V_M - 0.05, V_M + 0.05, 100)
V_out_tangent = (V_OH + V_OL) / 2 + g * (V_in_tangent - V_M)

# Plot the VTC curve
plt.figure(figsize=(10, 6))
plt.plot(V_in, V_out, label="VTC", color="blue", linewidth=2)

# Plot the tangent line at V_M
plt.plot(V_in_tangent, V_out_tangent, color="gray", linestyle="--", label="Slope at V_M")

# Plot V_M with a vertical line
plt.axvline(V_M, color="green", linestyle="--", label="V_M (Midpoint)")
# Plot V_OH and V_OL with horizontal lines
plt.axhline(V_OH, color="red", linestyle="--", label="V_OH (High Output)")
plt.axhline(V_OL, color="purple", linestyle="--", label="V_OL (Low Output)")
# Plot V_IL and V_IH with vertical lines
plt.axvline(V_IL, color="orange", linestyle="--", label="V_IL (Input Low)")
plt.axvline(V_IH, color="brown", linestyle="--", label="V_IH (Input High)")

# Annotate slope g at V_M
plt.text(V_M + 0.005, (V_OH + V_OL) / 2, f"Slope (g)  {g:.2f}", color="black", fontsize=12)

# Set plot labels and title
plt.xlabel("Input Voltage (V_in) [V]")
plt.ylabel("Output Voltage (V_out) [V]")
plt.title("Zoomed-In Voltage Transfer Characteristic (VTC) of CMOS Inverter")
plt.legend(loc="best")
plt.grid(True)

# Set x and y axis limits to zoom in
plt.xlim(0.6, 0.9)
plt.ylim(0.3, 1.2)

# Show the plot
plt.show()
```

```python
# Calculate the slope for the line from origin to the midpoint
slope_avg = (V_OH + V_OL) / (2 * V_M)

# Generate points for the line from the origin to (V_M, (V_OH + V_OL) / 2)
V_in_avg_line = np.linspace(0, V_M, 100)
V_out_avg_line = slope_avg * V_in_avg_line

# Plot the VTC curve, tangent line, and the new average line from the origin to V_M
plt.figure(figsize=(10, 6))
plt.plot(V_in, V_out, label="VTC", color="blue", linewidth=2)
plt.plot(V_in_tangent, V_out_tangent, color="gray", linestyle="--", label="Slope at V_M")
plt.plot(V_in_avg_line, V_out_avg_line, color="pink", linestyle=":", label="Line from Origin to V_M")

# Plot V_M with a vertical line
plt.axvline(V_M, color="green", linestyle="--", label="V_M (Midpoint)")
# Plot V_OH and V_OL with horizontal lines
plt.axhline(V_OH, color="red", linestyle="--", label="V_OH (High Output)")
plt.axhline(V_OL, color="purple", linestyle="--", label="V_OL (Low Output)")
# Plot V_IL and V_IH with vertical lines
plt.axvline(V_IL, color="orange", linestyle="--", label="V_IL (Input Low)")
plt.axvline(V_IH, color="brown", linestyle="--", label="V_IH (Input High)")

# Annotate slope g at V_M
plt.text(V_M + 0.005, (V_OH + V_OL) / 2, f"Slope (g)  {g:.2f}", color="black", fontsize=12)

# Set plot labels and title
plt.xlabel("Input Voltage (V_in) [V]")
plt.ylabel("Output Voltage (V_out) [V]")
plt.title("Zoomed-In Voltage Transfer Characteristic (VTC) of CMOS Inverter")
plt.legend(loc="best")
plt.grid(True)

# Set x and y axis limits to zoom in
plt.xlim(0, 0.9)  # Start at 0 to show the origin
plt.ylim(0, 1.2)  # Start at 0 to show the origin

# Show the plot
plt.show()
```
